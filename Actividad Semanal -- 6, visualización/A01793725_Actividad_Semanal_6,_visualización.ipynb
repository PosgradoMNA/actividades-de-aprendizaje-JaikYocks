*Count: Es el número de datos contenido en cada columna.

*Mean: Es el valor promedio del conjunto de datos por cada columna. 

*Std: Es la desviación estandar de cada conjunto de datos, que es una medida que se utiliza para cuantificar la variación o la dispersión de un conjunto de datos numéricos. Una desviación estándar baja indica que la mayor parte de los datos de una muestra tienden a estar agrupados cerca de su media (también denominada el valor esperado), mientras que una desviación estándar alta indica que los datos se extienden sobre un rango de valores más amplio.

*Los cuartiles son los tres valores –el primer cuartil en 25% (Q1), el segundo cuartil en 50% (Q2 o mediana) y el tercer cuartil en 75% (Q3)– que dividen una muestra de datos ordenados en cuatro partes iguales.

*min: El mínimo es el valor más pequeño de los datos. Se utiliza el mínimo para identificar un posible valor atípico o un error de entrada de datos. Una de las maneras más sencillas de evaluar la dispersión de los datos consiste en comparar el mínimo y el máximo. Si el valor mínimo es muy bajo, incluso cuando considere el centro, la dispersión y la forma de los datos, se investiga la causa del valor extremo.

*25%: Es el primer cuartil de la distribucion de datos (Q1), que dividen una muestra de datos ordenados en cuatro partes iguales junto con Q2 y Q3. El primer cuartil es el percentil 25 e indica que 25% de los datos es menor que o igual a este valor.

*50%: Es el segundo cuartil de la distribucion de datos (Q2 o mediana). que dividen una muestra de datos ordenados en cuatro partes iguales junto con Q1 y Q3.

*La mediana es el punto medio del conjunto de datos. El valor de este punto medio es el punto en el cual la mitad de las observaciones está por encima del valor y la otra mitad está por debajo del valor. La mediana se determina jerarquizando las observaciones y hallando la observación que ocupe el número [N + 1] / 2 en el orden jerarquizado. Si el número de observaciones es par, entonces la mediana es el valor promedio de las observaciones jerarquizadas en los números N / 2 y [N / 2] + 1.

Tanto la mediana como la media miden la tendencia central. Sin embargo, valores poco comunes, llamados valores atípicos, pueden afectar a la mediana menos de lo que afectan a la media. Si los datos son simétricos, la media y la mediana son similares.

*75%: Es el tercel cuartil de la distribucio de datos (Q3). que dividen una muestra de datos ordenados en cuatro partes iguales junto con Q1 y Q2. El tercer cuartil es el percentil que 75% de los datos es menor que o igual a este valor.

*max: El máximo es el valor más grande de los datos.

Se Utiliza el máximo para identificar un posible valor atípico o error de entrada de datos. Una de las maneras más sencillas de evaluar la dispersión de los datos consiste en comparar el mínimo y el máximo. 7. Reduce las dimensiones con PCA, si consideras necesario.

-Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades
de exploración de los datos la varianza > 70%. 

-Indica la importancia de las variables en cada componente.

```python
# Aplicamos PCA y normalización para poder trabajar con nuestra muestra

pcs = PCA()
pcs.fit(preprocessing.scale(ncredit.iloc[:, 0:14].dropna(axis=0))) 
```

```python
# Calculamos la desviación Estandar
# Calculamos la proporción de varianza
# Calculamos la proporción de varianza acumulativa

pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })

pcsSummary = pcsSummary.transpose()
pcsSummary.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary.columns) + 1)]
pcsSummary.round(4) 
```

A partir del componente 4 llegamos a un 72.01% de proporción de varianza acumulativa, por lo que 4 es el mínimo de componentes que representan la mayor variación de mis datos para un valor de varianza del 70%.

```python
# Se crea un dataframe con los 8 componentes principales
# Se crea un dataframe con los 8 componentes principales en valor absoluto

componentes_prin = pd.DataFrame(pcs.components_.transpose(), 
                                columns=pcsSummary.columns,
                                index=ncredit.iloc[:, 0:14].columns
                                )

valor_absoluto=pd.DataFrame(data=componentes_prin.abs())

componentes_prin.iloc[:,:4]
```

```python
print(valor_absoluto.iloc[:,:4],"\n")
```

```python
# componente principal PC1 en términos de las variables originales
print("El mayor contribuidor de PC1 es",valor_absoluto.PC1.nlargest(1),"\n")

print("El menor contribuidor de PC1 es",valor_absoluto.PC1.idxmin(),"\n")

# componente principal PC2 en términos de las variables originales
 
print("El mayor contribuidor de PC2 es",valor_absoluto.PC2.nlargest(1),"\n")

print("El menor contribuidor de PC2 es",valor_absoluto.PC2.idxmin(),"\n")

# componente principal PC3 en términos de las variables originales
 
print("El mayor contribuidor de PC3 es",valor_absoluto.PC3.nlargest(1),"\n")

print("El menor contribuidor de PC3 es",valor_absoluto.PC3.idxmin(),"\n")

# componente principal PC4 en términos de las variables originales
 
print("El mayor contribuidor de PC4 es",valor_absoluto.PC4.nlargest(1),"\n")

print("El menor contribuidor de PC4 es",valor_absoluto.PC4.idxmin(),"\n")
```

8. Elabora los histogramas de los atributos para visualizar su distribución.

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```
